```python
import pandas as pd

def process_regions(input_file, output_file):
    # Read the original CSV file without specifying column names
    df = pd.read_csv(input_file, header=None)

    # Remove any leading separator from the first cell
    df.iloc[:, 0] = df.iloc[:, 0].str.lstrip(' --- ')

    # Split the regions and create a DataFrame with the split regions
    regions_df = df.iloc[:, 0].str.split(' --- ', expand=True)

    # Determine the maximum number of regions
    max_regions = regions_df.shape[1]

    # Create column names region1, region2, ..., regionN
    column_names = [f'region{i}' for i in range(1, max_regions + 1)]

    # Rename the columns in the regions DataFrame
    regions_df.columns = column_names

    # Save the new DataFrame to a new CSV file
    regions_df.to_csv(output_file, index=False)

# Example usage:
input_csv_path = 'D://uw//data//main.csv'
output_csv_path = 'D://uw//data//clean.csv'

process_regions(input_csv_path, output_csv_path)
```

```python
import pandas as pd

def process_details(input_file, output_file):
    # Read the original CSV file
    df = pd.read_csv(input_file, header=None)

    # Filter out rows where the second column is an empty list
    df = df[df.iloc[:, 1].apply(lambda x: len(eval(x)) > 0)]

    # Split the second column by "\\n" to get key-value pairs
    key_value_pairs = df.iloc[:, 1].str.replace(r"\[|\]", "").str.split(r'\\n', expand=True)

    # Create a list of dictionaries to store the details
    details_list = []

    # Iterate through each row and assign values to columns
    for _, row in key_value_pairs.iterrows():
        details_dict = {}
        for item in row.dropna():
            key, value = map(str.strip, item.split(':', 1))
            details_dict[key] = value

        # Add the details to the list
        details_list.append(details_dict)

    # Create a DataFrame from the list of dictionaries
    details_df = pd.DataFrame(details_list)

    # Save the new DataFrame to a new CSV file
    details_df.to_csv(output_file, index=False)

# Example usage:
input_csv_path = 'D://uw//data//main.csv'
output_csv_path = 'D://uw//data//details.csv'

process_details(input_csv_path, output_csv_path)
```

```python
import csv
#               THIS IS PERFECT
place_name="HUANCAVELICA"


file_path = f'D://uw//data//{place_name}//details.txt'

import ast

# Read data from the text file and store it in the 'data' variable

with open(file_path, 'r', encoding='utf-8') as file:
    file_content = file.read()

# Add square brackets to the content to make it a valid list
file_content_with_brackets = f"[{file_content}]"

# Safely evaluate the string as a Python literal (list of dictionaries)
data = ast.literal_eval(file_content_with_brackets)
# Extract all unique keys
all_keys = set()
for entry in data:
    all_keys.update(entry.keys())

# Sort the keys for consistent column order in the CSV
sorted_keys = sorted(all_keys)

# Write to CSV file
csv_file_path = f'D://uw//data//{place_name}//clean.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=sorted_keys)
    
    # Write header
    csv_writer.writeheader()
    
    # Write rows
    for entry in data:
        row_data = {key: entry.get(key, '') for key in sorted_keys}
        csv_writer.writerow(row_data)

print(f"clean file has been created at: {csv_file_path}")


file_path = f'D://uw//data//{place_name}//region.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    data = [line.strip(" ',\n") for line in file.readlines()]

# Remove the last string after "---" in each row
cleaned_rows = [row.rsplit('---', 1)[0].strip() for row in data]

# Split the remaining string using "---" and determine the maximum number of regions
split_rows = [row.split('---') for row in cleaned_rows]
max_regions = max(len(row) - 1 for row in split_rows)

# Create CSV file with columns named "Region 1", "Region 2", ..., "Region n"
csv_file_path = f'D://uw//data//{place_name}//regions_list.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = [f"Region {i}" for i in range(1, max_regions + 1)]
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    # Write header
    csv_writer.writeheader()
    
    # Write rows
    for row in split_rows:
        row_data = {f"Region {i + 1}": region.strip() if i < len(row) - 1 else '' for i, region in enumerate(row[1:])}
        csv_writer.writerow(row_data)

print(f"region_list file has been created at: {csv_file_path}")
file_path = f'D://uw//data//{place_name}//region.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    data = [line.strip(" ',\n") for line in file.readlines()]

rows = [row.strip().split('---')[-1].strip() for row in data]

# Write to CSV file
csv_file_path = f'D://uw//data//{place_name}//company.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # Write rows
    csv_writer.writerows([[row] for row in rows])

print(f"company CSV file has been created at: {csv_file_path}")
```

```python


from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--headless') 
options = webdriver.ChromeOptions()
url='https://www.universidadperu.com/empresas/cultivo-de-hortalizas-y-legumbres-categoria.php'


def append_to_csv(file_path, data_list1, data_list2):
    """
    Appends two lists of key-value pairs to a CSV file.

    Parameters:
    - file_path (str): Path to the CSV file.
    - data_list1 (list): First list of key-value pairs.
    - data_list2 (list): Second list of key-value pairs.

    Returns:
    - None
    """
    with open(file_path, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        
        # Use zip to iterate over both lists simultaneously
        for item1, item2 in zip(data_list1, data_list2):
            # Write each pair as a row in the CSV file
            csv_writer.writerow([item1, item2])

# Specify the file path


def extract_data(html_content):
    """
    Extracts structured data from HTML content.

    Parameters:
    - html_content (str): HTML content of a page.

    Returns:
    - List of strings: Each string contains key-value pairs separated by ":" and each pair on a new line.
    """
    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')

    # Initialize a list to store the structured data
    result_list = []

    # Find the main <ul> elements
    ul_elements = soup.find_all('ul')

    # Process each <ul> element
    for ul_element in ul_elements:
        ul_data = {}  # Dictionary to store data for the current <ul> element
        li_elements = ul_element.find_all('li')
        for li_element in li_elements:
            strong_tag = li_element.find('strong')
            if strong_tag:
                key = strong_tag.text.strip(':')
                # Check if there is content after the <strong> tag
                value = li_element.contents[1].strip() if len(li_element.contents) > 1 else ""

                # Check for <span> elements within <li>
                span_element = li_element.find('span')
                if span_element:
                    span_key = f"{key} (Span)"
                    span_value = span_element.text.strip()
                    if span_value and key != span_value:  # Exclude empty and identical span values
                        ul_data[span_key] = span_value

                if value and key != value:  # Exclude empty and identical values
                    ul_data[key] = value

        # Convert the data to a string and append to the result_list
        ul_data_string = '\n'.join([f"{key}: {value}" for key, value in ul_data.items()])
        # Check if the data is not a duplicate
        if ul_data_string not in result_list:
            result_list.append(ul_data_string)

    return result_list

def clean_data(html_data):
    soup = BeautifulSoup(html_data, 'html.parser')
    data_dict = {}

    # Find all 'ul' elements
    ul_elements = soup.find_all('ul')

    for ul in ul_elements:
        # Find all 'li' elements within each 'ul'
        list_items = ul.find_all('li')

        for item in list_items:
            # Extract text content from each 'li' element
            item_text = item.get_text(separator=' ')

            # Split the text into key and value using the colon
            parts = item_text.split(':')

            if len(parts) == 2:
                key = parts[0].strip()
                value = parts[1].strip()

                # Add key-value pair to the dictionary
                data_dict[key] = value

    return data_dict

def loop(link, name, root,proxy_address,proxy_port):
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--headless')

    proxy_url = f'http://{proxy_address}:{proxy_port}'
    #options.add_argument(f'--proxy-server={proxy_url}')
    
    current_root = root + " --- " + name  # Concatenate the current name to the root
    print(f"root: {current_root}")
    url = link
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    time.sleep(2)
    ul_element = driver.find_element('tag name', 'ul')
    li_elements = ul_element.find_elements('tag name', 'li')
    flag = 1

    try:
        for li_element in li_elements:
            text_content = li_element.text
            link_element = li_element.find_element('tag name', 'a')
            link_url = link_element.get_attribute('href')
    except:
        flag = 0
    if flag:
        #print("found links")
        for li_element in li_elements:
            text_content = li_element.text
            #print(f"Text Content: {text_content}")
            link_element = li_element.find_element('tag name', 'a')
            link_url = link_element.get_attribute('href')
            #print(f"Link Content: {link_url}")

            loop(link_url, text_content, current_root,proxy_address,proxy_port)  # Pass the current_root to the recursive call
    else:
          
        target_div = driver.find_element(By.XPATH,'//div[@itemscope][@itemtype="https://schema.org/LocalBusiness"]')

        ul_elements = target_div.find_elements('tag name', 'ul')

        # Print the contents of each <ul> element
        div_content=""
        for ul_element in ul_elements:
            ul_content = ul_element.get_attribute("outerHTML")
            div_content+=ul_content

        details.append(clean_data(div_content))
        
        #append_to_csv(csv_file_path, region, details)
        #print("csv updated")
            
        region.append(current_root)
            
        '''
        print("--------------------DONE----------------------")
        print(t)
        print("--------------------DONE----------------------")'''
    driver.quit()
region=[]
companies=[]
details=[]

def my_main(proxy_address,proxy_port):
    options = webdriver.ChromeOptions()
    

    proxy_url = f'http://{proxy_address}:{proxy_port}'
    #options.add_argument(f'--proxy-server={proxy_url}')
    
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--headless')

    url='https://www.universidadperu.com/empresas/cultivo-de-hortalizas-y-legumbres-categoria.php'
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    time.sleep(2)
    ul_element = driver.find_element('tag name', 'ul')
    li_elements = ul_element.find_elements('tag name', 'li')
    
    for li_element in li_elements:
        text_content = li_element.text
        print(f"Text Content: {text_content}")
        link_element = li_element.find_element('tag name', 'a')
        link_url = link_element.get_attribute('href')
        print(f"Link URL: {link_url}")
        if text_content=="LAMBAYEQUE":
            loop(link_url,text_content,"",proxy_address,proxy_port)
            break



```
